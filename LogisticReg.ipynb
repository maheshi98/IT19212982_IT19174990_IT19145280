# Data Wrangling

```python
weatherData.isnull()
```

```python
weatherData.drop(['Cloud3pm' , 'Cloud9am'] , axis =1 , inplace = True)
```

```python
weatherData.head(10)
```

```python
weatherData.drop(['Evaporation' , 'Sunshine'] , axis =1 , inplace = True)
```

```python
weatherData.head(10)
```

```python
weatherData.dropna(subset =['RainTomorrow' , 'RainToday'])
```

```python
weatherData.dropna(inplace = True)
```

```python
sns.heatmap(weatherData.isnull() , yticklabels = False , cbar = False)
weatherData.isnull().sum()
```

```python
numaricData = []
categoricData = []

for column in weatherData:
    if column != 'RainTomorrow':
        if is_numeric_dtype(weatherData[column]):
            numaricData.append(column)
        elif is_string_dtype(weatherData[column]):
            categoricData.append(column)
print("Numarical data :",numaricData)
print("Categorical Data :",categoricData)
```

```python
weatherData.fillna(weatherData.mean() , inplace =True)
weatherData.head()
```

```python
for i in (categoricData):
    if weatherData[i].isnull().any():
        weatherData[i].fillna("UW")
```

```python
weatherData.head()
```

# Feature Engineering 

```python
#Automate a for loop for univariate analysis
#for a numerical variable a histogram will generate to visualize data
#distribution
#for a categorical variable a bar chart will generate to visualize data
#distribution

for column in weatherData:
    mplt.figure(column, figsize=(5,5))
    mplt.title(column)
    if is_numeric_dtype(weatherData[column]):
        weatherData[column].plot(kind='hist')
    elif is_string_dtype(weatherData[column]):
        #show onlytop10values in each categoricaldata
        weatherData[column].value_counts()[:10].plot(kind='bar')
    
```

```python
#address outliers in 'Rainfall'
#use quantile(0.9) to limit the dataset
maxData = weatherData['Rainfall'].quantile(0.9)
weatherData = weatherData[weatherData["Rainfall"] < maxData]
weatherData["Rainfall"].plot(kind='hist')
weatherData.shape
```

```python
#Feature Transformation
#Transform 'Date' variable to 'Month'
weatherData['Month']= panda.to_datetime(weatherData['Date']).dt.month.apply(str)
weatherData['Month'].value_counts().plot(kind='bar')
```

```python
#Categorical Feature Encoding
#use label encoding
from sklearn.preprocessing import LabelEncoder

catFeatures = ['Location','WindGustDir','WindDir9am','WindDir3pm','RainToday','Month','RainTomorrow']

for n in catFeatures:
    weatherData[n] = LabelEncoder().fit_transform(weatherData[n])
```

```python
#Feature selection
#use correlation analysis
mplt.figure(1,figsize=(15,15))
correlationValue = weatherData.corr()
sns.heatmap(correlationValue,cmap = "GnBu",annot = True)
```
